# seaborn: KDEPlot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg").dropna()
print(cars.shape)
```

```python
print(cars.head())
```

# Univariate(1D) plot
It means plot contains only one variable

```python
sns.set_style('whitegrid')
sns.kdeplot(cars.horsepower)
```

```python
sns.kdeplot(cars.horsepower, shade=True)
```

```python
sns.kdeplot(cars.horsepower, bw_adjust=10)
```

```python
sns.kdeplot(cars.horsepower, cumulative=True)
```

# Bivariate(2D) Plot
It means plot contain two variable
Join Probablity

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=20)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=True)
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=True)
```
